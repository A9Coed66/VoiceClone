```python
# Đọc và hiển thị dữ liệu từ file wav
import wave
import matplotlib.pyplot as plt
import numpy as np
```

```python
wfobj = wave.open("vo_hutao_draw_appear.wav")
# read all data
data = wfobj.readframes(-1)
# convert into numpy struct
signal = np.frombuffer(data, "int16")
# normalize into [-1, 1]
signal = signal/abs(signal).max()
plt.plot(signal)
plt.title('waveform (time-domain)')
plt.grid(True)
plt.show()
```

```python
# Thực hiện quan sát phổ của tín hiệu
wfobj = wave.open("vo_hutao_draw_appear.wav", 'rb')
# read all data
data = wfobj.readframes(-1)
signal = np.frombuffer(data, "int16")
seg = signal
seg = seg/abs(seg).max()
Y = np.fft.rfft(seg)/len(seg)
Y = Y[range(int(len(seg)/2))]
plt.figure()
# plt.subplot(211)
plt.grid(True)
plt.title("waveform (time-domain)")
# plt.subplot(212)
plt.semilogy(abs(Y))
plt.grid(True)
plt.title('spectrum')
plt.show()
```
